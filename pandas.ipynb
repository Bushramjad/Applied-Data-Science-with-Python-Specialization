**Reading CSV into Dataframe and basic dataframe operations**

```python
# reading csv into dataframe
df = pd.read_csv('/content/drive/MyDrive/resources/week-1/datasets/Admission_Predict.csv')
df.head()
```

```python
# reading csv without creating dataframe index
df = pd.read_csv('/content/drive/MyDrive/resources/week-1/datasets/Admission_Predict.csv', index_col=0)
df.head()
```

```python
# renaming column names
df_new = df.rename(columns={'SOP' : 'Statement of purpose',
                            'LOR ' : 'Letter of Recommendation'})
```

```python
df_new.head()
```

```python
#changing all of the column name to lower case
cols = list(df.columns)
```

```python
cols = [x.lower().strip() for x in cols]
df.columns = cols
df.head()
```

**Boolean Masking**

```python
#boolean mask and broadcasting
admit_mask = df['chance of admit'] > 0.7
admit_mask.head()
```

```python
df.where(admit_mask).head()
```

```python
# now we want to get rid of NaN rows
df.where(admit_mask).dropna().head()
```

```python
# alternate way (where() and dropna() combined)
df[df['chance of admit'] > 0.7].head()
```

```python
# multiple boolean masks ( and -> &, or -> |)
((df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)).head()
```

**Indexing Dataframes**

```python
df.head()
```

```python
# we want 'chance of admit' as an index, preserving serial no. in another column for future use
df['Serial Number'] = df.index
df = df.set_index('chance of admit')
df.head()
```

```python
# resetting index
df = df.reset_index()
df.head()
```

```python
# how to query this dataframe?
# using loc attribute
# for single row
df.loc['Michigan' , 'Washtenaw County']
```
